```python
import pandas as pd
import tsfel
```

```python
data_path = "../data/raw/sp500_top10_stocks_raw.csv"
df = pd.read_csv(data_path)

# 2. Tiền xử lý: Chuyển Date về dạng datetime và sắp xếp theo trình tự thời gian
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by=['Ticker', 'Date'])


cfg = tsfel.get_features_by_domain() 

def extract_features_for_ticker(group):
    data_to_extract = group[['Open', 'High', 'Low', 'Close', 'Volume']]

    features = tsfel.time_series_features_extractor(cfg, data_to_extract, window_size=20)
    
    features['Ticker'] = group['Ticker'].iloc[0]
    return features

all_ticker_features = []

for ticker, group in df.groupby('Ticker'):
    print(f"Processing features for: {ticker}")
    ticker_feat = extract_features_for_ticker(group)
    all_ticker_features.append(ticker_feat)

final_features_df = pd.concat(all_ticker_features, ignore_index=True)

display(final_features_df)
```
