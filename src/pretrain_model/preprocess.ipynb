```python
import pandas as pd
import glob
import warnings
warnings.filterwarnings('ignore')
```

```python
stock_price_folder = "f:/huyle_data_projects/stock_price_prediction/data/for_pretrained_models/prices"
macro_folder = "f:/huyle_data_projects/stock_price_prediction/data/for_pretrained_models/macro"
```

```python
stock_files = glob.glob(f"{stock_price_folder}/*.csv")
stock_list = []

for f in stock_files:
    df = pd.read_csv(f)
    df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%y')
    df = df[df['Date'] >= '2010-01-01'] # filter data from 2010 onwards
    stock_list.append(df)
combined_stocks = pd.concat(stock_list, ignore_index=True)


rename_map = {
    'DX-Y.NYB': 'dxy',
    'FEDFUNDS': 'fed_rate',
    'GC=F': 'gold',
    'CL=F': 'oil',
    '^GSPC': 'sp500'
}

macro_files = glob.glob(f"{macro_folder}/*.csv")
macro_list = []

for f in macro_files:
    df = pd.read_csv(f)
    df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%y')
    # df = df[df['Date'] >= '2010-01-04'] # filetr data from 04-01-2010 onwards to match the data from stock price files
    ticker = df['Ticker'].iloc[0]
    col_name = rename_map.get(ticker, ticker)
    df = df[['Date', 'Value']].rename(columns={'Value': col_name})
    df.set_index('Date', inplace=True)
    macro_list.append(df)

combined_macro = pd.concat(macro_list, axis=1).reset_index()
combined_macro = combined_macro.sort_values('Date').reset_index(drop=True)
```

## Join 2 dataframes

```python
# merge using left join to keep vn trading days as the primary timeline
combined_df = pd.merge(combined_stocks, combined_macro, on='Date', how='left')

# ffill macro data within each ticker group to handle missing dates after merge
macro_cols = ['dxy', 'fed_rate', 'gold', 'oil', 'sp500']
combined_df[macro_cols] = combined_df.groupby('Ticker')[macro_cols].ffill()

combined_output_path = "f:/huyle_data_projects/stock_price_prediction/data/for_pretrained_models/processed/combined.csv"
combined_df.to_csv(combined_output_path, index=False)
```
